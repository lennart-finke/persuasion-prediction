```python
import json
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
import statsmodels.api as sm
import scipy
from persuasion import get_persuasion_data

"""
A notebook to analyze the correlations of model persuasiveness predictions of human agreement change ("persuasiveness metric). Change the FILENAMES variable below (logs/XYZ.json -> XYZ) to take in logs from Inspect runs from persuasion.py. 
"""
FILENAMES = ["gpt3-5", "gpt4o-mini", "gpt4o"]

for filename in FILENAMES:
    with open("logs/" + filename + ".json", "r", encoding="utf-8") as fp:
        log = json.load(fp)

    df = get_persuasion_data()
    for col in ["rating_initial", "rating_final"]:
        df[col] = [float(x[0]) for x in df[col]]

    samples = log["samples"]

    n_rows = df.shape[0]
    predicted = n_rows*[0]
    target = n_rows*[0]
    for sample in samples:
        id = int(sample["id"])
        target[id] = int(sample["target"])
        assert target[id] == df["rating_final"].iloc[id]
        predicted[id] = float(sample["scores"]["absolute_difference"]["answer"])

    df["predicted"] = predicted
    df = df[df["predicted"] != 0]
    df = df.sample(795, random_state=42)
    target = df["rating_final"] - df["rating_initial"]
    predicted = df["predicted"] - df["rating_initial"]

    num_correct = sum(1 for x,y in zip(predicted, target) if x-1==y)
    print(f"Accuracy: {num_correct / len(target)}")
     
    x = predicted
    y = target

    x_min, x_max = x.min(), x.max()
    y_min, y_max = y.min(), y.max()
    x_bins = np.arange(x_min, x_max + 2) - 0.5
    y_bins = np.arange(y_min, y_max + 2) - 0.5

    plt.hist2d(x, y, bins=[x_bins, y_bins])
    plt.xlabel(f'Predicted Human Persuasiveness ({filename})')
    plt.ylabel(f'Ground Truth Persuasiveness')
    plt.title(f'Heatmap of Value Pair Frequencies: {filename}')
    print(Counter(zip(target, predicted)))
    X = sm.add_constant(x) 
    y = y
    model = sm.OLS(y, X).fit()
    print(model.summary())
    res = scipy.stats.linregress(predicted, target)
    print(res)

    reg_predictions = np.round(res.intercept + res.slope * predicted)
    num_correct = sum(1 for x,y in zip(reg_predictions, target) if x==y)
    print(f"Regression Accuracy: {num_correct / len(target)}")

    plt.plot(range(int(min(x)),int(max(x))+1), res.intercept + res.slope*range(int(min(x)),int(max(x))+1), 'r', label='fitted line')
    plt.gca().set_aspect('equal')
    plt.show()
```
